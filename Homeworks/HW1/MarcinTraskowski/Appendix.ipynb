```python
columns = ["Age", "Workclass", "Final Weight", "Education", "Education Number of Years", "Marital-status", "Occupation", "Relationship", "Race", "Gender", "Capital-gain", "Capital-loss", "Hours-per-week", "Native-country", "Income"]
df = pd.read_csv("datasets/adult/raw/adult.data", index_col=None, header=None, names=columns)

df.replace(" ?", np.nan, inplace=True)
df = df.dropna()

df.iloc[:,[0,2,4,10,11,12]] = df.iloc[:,[0,2,4,10,11,12]].astype(np.float64)

df['Income'] = df['Income'].map({' <=50K': 0, " >50K": 1})
df.iloc[:,14] = df.iloc[:,14].astype(np.int64)
print(df.shape)
df.head()
```

```python
# we already have division for train and test set
# i will briefly join them and then one hot encode and then divide again in the same proportion

df_combined = pd.concat([df, df_test], ignore_index=True)
print(df_combined.shape)
```

```python
categorical_cols = df_combined.columns[[1, 3, 5, 6, 7, 8, 9, 13]]
df_combined = pd.get_dummies(df_combined, columns=categorical_cols, drop_first=True)

# Placing 'Income' at the end
income_column = df_combined.pop('Income')
df_combined['Income'] = income_column
```

```python
X_train = df_combined.iloc[0:30162,0:96]
y_train = df_combined.iloc[0:30162,96]
X_test = df_combined.iloc[30162:,0:96]
y_test = df_combined.iloc[30162:,96]

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# this is function for calculating 
# Demographic Parity, Equal Opportunity, Predictive Rate Parity
from sklearn.metrics import confusion_matrix

def Fairness_Calculation(model, X_test, y_test, attribute):

    X_test_Group_0 = X_test[X_test[attribute] == 0] 
    X_test_Group_1 = X_test[X_test[attribute] == 1]

    y_test_Group_0 = y_test[X_test[attribute] == 0]
    y_test_Group_1 = y_test[X_test[attribute] == 1]

    y_pred_Group_0 = model.predict(X_test_Group_0)
    y_pred_Group_1 = model.predict(X_test_Group_1)

    # Demographic Parity
    
    Demographic_Parity_0 = np.mean(y_pred_Group_0)
    Demographic_Parity_1 = np.mean(y_pred_Group_1)
    
    
    TN_0, FP_0, FN_0, TP_0 = confusion_matrix(y_true = y_test_Group_0, y_pred = y_pred_Group_0).ravel()
    TN_1, FP_1, FN_1, TP_1 = confusion_matrix(y_true = y_test_Group_1, y_pred =  y_pred_Group_1).ravel()

    # Equal Opportunity
    
    Equal_Opportunity_0 = (TP_0 / (TP_0 + FN_0))
    Equal_Opportunity_1 = (TP_1 / (TP_1 + FN_1))
    

    # Predictive Rate Parity
    
    Predictive_Rate_Parity_0 = (TP_0 / (TP_0 + FP_0))
    Predictive_Rate_Parity_1 = (TP_1 / (TP_1 + FP_1))

    return (Demographic_Parity_0, Demographic_Parity_1,
            Equal_Opportunity_0, Equal_Opportunity_1,
            Predictive_Rate_Parity_0, Predictive_Rate_Parity_1
           )
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

RandomForest = RandomForestClassifier(n_estimators=100, random_state=120)

RandomForest.fit(X_train, y_train)

y_pred_RandomForest = RandomForest.predict(X_test)

print("\n\nFirst Model Accuracy:", accuracy_score(y_true = y_test, y_pred =  y_pred_RandomForest))

Fairness_values = Fairness_Calculation(RandomForest, X_test, y_test, 'Gender_ Male')


print("\nDemographic_Parity_0 : ",Fairness_values[0])
print("\nDemographic_Parity_1 : ",Fairness_values[1])
print("\nDemographic_Parity Rate : ",round(Fairness_values[0]/Fairness_values[1],2))

print("\n\nEqual_Opportunity_0 : ",Fairness_values[2])
print("\nEqual_Opportunity_1 : ",Fairness_values[3])
print("\nEqual_Opportunity Rate : ",round(Fairness_values[2]/Fairness_values[3],2))

print("\n\nPredictive_Rate_Parity_0 : ",Fairness_values[4])
print("\nPredictive_Rate_Parity_1 : ",Fairness_values[5])
print("\nPredictive_Rate_Parity Rate : ",round(Fairness_values[4]/Fairness_values[5],2))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

MLP = MLPClassifier(hidden_layer_sizes=(32, 32,32), max_iter=200,verbose = False, random_state=120)

MLP.fit(X_train, y_train)

y_pred_MLP = MLP.predict(X_test)

print("\n\nSecond Model Accuracy:", accuracy_score(y_true = y_test, y_pred =  y_pred_MLP))

Fairness_values = Fairness_Calculation(MLP, X_test, y_test, 'Gender_ Male')


print("\nDemographic_Parity_0 : ",Fairness_values[0])
print("\nDemographic_Parity_1 : ",Fairness_values[1])
print("\nDemographic_Parity Rate : ",round(Fairness_values[0]/Fairness_values[1],2))

print("\n\nEqual_Opportunity_0 : ",Fairness_values[2])
print("\nEqual_Opportunity_1 : ",Fairness_values[3])
print("\nEqual_Opportunity Rate : ",round(Fairness_values[2]/Fairness_values[3],2))

print("\n\nPredictive_Rate_Parity_0 : ",Fairness_values[4])
print("\nPredictive_Rate_Parity_1 : ",Fairness_values[5])
print("\nPredictive_Rate_Parity Rate : ",round(Fairness_values[4]/Fairness_values[5],2))
```

```python
# Bias mitigation: Data Pre-processing, subsambling

from imblearn.under_sampling import RandomUnderSampler
index = X_train.columns.get_loc('Gender_ Male')

y = X_train['Gender_ Male']

X = pd.concat([ X_train.drop(columns=['Gender_ Male']), y_train], axis =1 )

smote = RandomUnderSampler(sampling_strategy='all', random_state=1304)

X_Resampled, y_Resampled = smote.fit_resample(X, y)


print("\n\nOriginal distribution of Gender_ Male (IN TRAINING DATA):")
print(X_train['Gender_ Male'].value_counts())

print("\nResampled distribution of Gender_ Male:")
print(y_Resampled.value_counts())

X_Resampled.insert(index, 'Gender_ Male', y_Resampled.values)

print(X_Resampled.shape)
```

```python
X_train_Resampled = X_Resampled.iloc[:,0:96]
y_train_Resampled = X_Resampled.iloc[:,96]

RandomForest_Resampled = RandomForestClassifier(n_estimators=100, random_state=120)

RandomForest_Resampled.fit(X_train_Resampled, y_train_Resampled)

y_pred_RandomForest_Resampled = RandomForest_Resampled.predict(X_test)

print("\n\nThird Model Accuracy:", accuracy_score(y_true = y_test, y_pred =  y_pred_RandomForest_Resampled))

Fairness_values = Fairness_Calculation(RandomForest_Resampled, X_test, y_test, 'Gender_ Male')


print("\nDemographic_Parity_0 : ",Fairness_values[0])
print("\nDemographic_Parity_1 : ",Fairness_values[1])
print("\nDemographic_Parity Rate : ",round(Fairness_values[0]/Fairness_values[1],2))

print("\n\nEqual_Opportunity_0 : ",Fairness_values[2])
print("\nEqual_Opportunity_1 : ",Fairness_values[3])
print("\nEqual_Opportunity Rate : ",round(Fairness_values[2]/Fairness_values[3],2))

print("\n\nPredictive_Rate_Parity_0 : ",Fairness_values[4])
print("\nPredictive_Rate_Parity_1 : ",Fairness_values[5])
print("\nPredictive_Rate_Parity Rate : ",round(Fairness_values[4]/Fairness_values[5],2))
```
